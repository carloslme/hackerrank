# Solution
**Query**
```bash
SELECT
    DISTINCT(CITY)
FROM STATION
WHERE 1=1
    AND LEFT(CITY, 1) IN ('a','e','i','o','u')
    AND RIGHT(CITY, 1) IN ('a','e','i','o','u');
```
