# Solution

**Query**

```bash
  WITH min_len AS (
    SELECT 
      CITY
      , LENGTH(CITY)
    FROM STATION
    ORDER BY 2, 1 ASC
    LIMIT 1
  )
  ,
  max_len AS (
    SELECT
      CITY
      , LENGTH(CITY)
    FROM STATION
    ORDER BY 2 DESC
  )
  (SELECT * FROM min_len)
  UNION
  (SELECT * FROM max_len)
```

**Output**

```bash
Amo 3
Marine On Saint Croix 21
```
